Concatenating Data Frames 

```python
# Create the first DataFrame with random numbers
first = pd.DataFrame(np.random.randn(5, 4))

# Create the second DataFrame with random numbers
second = pd.DataFrame(np.random.randn(5, 4))

# Combine (concatenate) the two DataFrames by stacking them vertically
combined = pd.concat([first, second])

# Display the combined DataFrame
print(combined)
```

```python
# INNER JOIN: Combines data from two tables only where there is a match in the key column

# Create the first DataFrame
left = pd.DataFrame({
    'netID': ['adg133', 'abc134', 'def234'],
    'midterm': [67, 89, 90]
})

# Create the second DataFrame
right = pd.DataFrame({
    'netID': ['adg133', 'abc134', 'def234'],
    'finals': [87, 49, 78]
})

# Merge the two DataFrames on the 'netID' column
merged = pd.merge(left, right, on='netID')

# Display the merged DataFrame
print(merged)
```

```python
# OUTER JOIN: Combines data from two tables and includes all rows from both tables, filling in NaN (or NULL) where data is missing.

left = pd.DataFrame({
    'netID': ['adg133', 'abc123'],
    'midterm': [65, 45]
})

# Create the second DataFrame
right = pd.DataFrame({
    'netID': ['adg133'],
    'final': [80]
})

# Concatenate the two DataFrames
common = pd.concat([left, right])

# Display the combined DataFrame
print(common)
```

```python
#LEFT JOIN: keeps all rows from the left table, and only the matching rows from the right table. If there is no match, it fills with NaN (or NULL).

# Create the left DataFrame
left = pd.DataFrame({
    'netID': ['adg133', 'abc123'],
    'midterm': [65, 45]
})

# Create the right DataFrame
right = pd.DataFrame({
    'netID': ['adg133'],
    'final': [80]
})

# Perform a left join on the two DataFrames
common = pd.merge(left, right, how="left")

# Display the result
print(common)
```

```python
#RIGHT JOIN: keeps all rows from the right table, and only the matching rows from the left table. If there is no match, it fills with NaN (or NULL).

# Create the left DataFrame
left = pd.DataFrame({
    'netID': ['adg133', 'abc123'],
    'midterm': [65, 45]
})

# Create the right DataFrame
right = pd.DataFrame({
    'netID': ['adg133'],
    'final': [80]
})

# Perform a right join on the two DataFrames
common = pd.merge(left, right, how="right")

# Display the result
print(common)
```

Memory in Pandas 

Slide 1: Internal Representation of Numeric and String Data
 -Numeric Data: Stored as blocks of numbers, similar to an array, with pointers to where the data is stored.
 -String Data: More complex. Each string has a "pointer" leading to the actual characters stored in memory. This means strings use more memory and take longer to process compared to numeric data.

Slide 2: Internal Representation of a DataFrame
 -DataFrame Layout: A DataFrame is split into blocks based on data types:
 -IntBlock: Stores all integer data.
 -ObjectBlock: Stores all string or mixed-type data.
 -FloatBlock: Stores all float data.
 -Efficiency: Grouping data by type helps Pandas optimize performance. Numeric blocks are faster and use less memory compared to object blocks.

Slide 3: Memory Usage by Pandas DataFrames
 -Memory Usage Overview:
  -Pandas estimates how much memory the DataFrame uses.
  -For higher accuracy, use memory_usage(deep=True) to include deeper memory (e.g., strings).
 -Memory Usage by Data Type:
  -Numeric data (int, float) uses less memory.
  -String data (object type) uses more memory because of pointers and variable lengths.
  -The output shows average memory usage for each data type.


In short:
 -Numeric data is compact and efficient.
 -Strings (object type) consume more memory and are slower to process.
 -Pandas organizes DataFrames into blocks by data type to improve performance.

```python
# Downcasting, converting a variable or DataFrame column into a smaller or more memory-efficient data type without losing information, done to reduce memory usage
# Use pd.to_numeric to down cast numeric types 

# Credit: Josh Devlin
def mem_usage(pandas_obj):
    if isinstance(pandas_obj, pd.DataFrame):
        usage_b = pandas_obj.memory_usage(deep=True).sum()
    else:  # we assume if not a df it's a series
        usage_b = pandas_obj.memory_usage(deep=True)
    usage_mb = usage_b / 1024 ** 2  # convert bytes to megabytes
    return "{:03.2f} MB".format(usage_mb)

df_int = df.select_dtypes(include=['int'])
converted_int = df_int.apply(pd.to_numeric, downcast='unsigned')

print(mem_usage(df_int))
print(mem_usage(converted_int))

compare_ints = pd.concat([df_int.dtypes, converted_int.dtypes], axis=1)
compare_ints.columns = ['before', 'after']
compare_ints.apply(pd.Series.value_counts)
```

```python
# Change in memory usage 

df_float = df.select_dtypes(include=['float'])
converted_float = df_float.apply(pd.to_numeric, downcast='float')

print(mem_usage(df_float))
print(mem_usage(converted_float))

# This will show that floats can be downcasted from 64 bit to 32-bit (50% savings)
compare_floats = pd.concat([df_float.dtypes, converted_float.dtypes], axis=1)
compare_floats.columns = ['before', 'after']
compare_floats.apply(pd.Series.value_counts)
```
